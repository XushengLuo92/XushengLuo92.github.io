```python
from scipy.io import loadmat

data = loadmat('/Users/xushengluo/Downloads/mpii_human_pose_v1_u12_2/mpii_human_pose_v1_u12_1.mat')
print(data.keys())
```

```python
R = data['RELEASE']
annolist = R['annolist'][0, 0]
single_person = R['single_person'][0, 0]

# Collect image names with exactly one single person
imgs_with_one_single_person = []

for i in range(annolist.shape[1]):  # iterate over all images
    sp = single_person[i]
    if sp.size > 1:  # exactly one isolated person
        name = annolist[0, i]['image'][0, 0]['name'][0]
        imgs_with_one_single_person.append(name)

print("Number of single-person images:", len(imgs_with_one_single_person))
print("First few:", imgs_with_one_single_person[:10])
```
